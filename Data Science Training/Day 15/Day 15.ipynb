```python
data = {'Study_hrs':range(1,11),
       'Marks Obtained':[25,33,46,50,55,67,78,85,95,100]}

df = pd.DataFrame(data)
df
```

```python
# ML model build?
# Scatter Plot
plt.title('Study vs Marks')
plt.scatter(df['Study_hrs'],df['Marks Obtained'],marker='x', color='r')
plt.xlabel('Study Hrs')
plt.ylabel('Marks')
plt.show()
```

```python
# Training : Zero Error
# Future: Predict:
```

```python
# # Y = mx + c
# # M : Slope
# # C : Intercept

# Y = Marks
# x = Study_hrs
```

```python
# x(1), y(1): simple Linear Regression
# x(multiple): y(1): Multiple Linear Regression

df
```

```python
# Import Module
from sklearn.linear_model import LinearRegression

# Object Model Creation
model = LinearRegression()

# X, y

X = df[['Study_hrs']]
y = df['Marks Obtained']

# Training/Learning
model.fit(X,y)
```

```python
x = [[5.5]]

m = model.coef_
c = model.intercept_

print(m,c)
```

```python
#  y = m*x + c
std_hrs = [[5.5]]

# "y =  + x"
# y = wx +  c

marks = m * std_hrs + c

print('Predicted Marks:',marks[0][0])
```

```python
import warnings
warnings.filterwarnings('ignore')
study_hrs = input('Enter study hours')
std_hrs = [[float(study_hrs)]]

ans = model.predict(std_hrs)
marks = round(ans[0],2)

if marks > 100:
    marks = 100
elif marks < 0 :
    marks = 0


print(f"Your Predicted Marksis :{marks}")
```

```python
# Description:
# The advertising dataset captures the sales revenue generate
# with respect to advertisement costs across multiple channels like radio, tv, and newspapers.
# It is required to understand the impact of ad budgets on the overall sales.
```

```python
# y = m*x + c

# X = 3 features

# y = m1*x1 + m2*x2 + m3*x3 + c

# y =  + 1*x1 + 2*x2 + 3*x3

# y = w1*x1 + w2*x2 + m3*x3 + c

# x1 = TV_budget
# x2 = radio
# x3 = newspaper
```

```python
X = df_multi.iloc[:,:-1]
y = df_multi['Sales']
```

```python
model = LinearRegression()

model.fit(X,y)
```

```python
tv = 100
radio = 50
newspaper = 100
budget = [[tv,radio,newspaper]]

pre_sales = round(model.predict(budget)[0],2)
print('Predicted Sales:',pre_sales)
```
