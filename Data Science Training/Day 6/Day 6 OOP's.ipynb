```python
class ATM:
    def __init__(self):
        print('Object created successfully')
        self.pin = 1111   # public variable
        self.balance = 2000  # public variable

    def change_pin(self):  # public method
        self.pin = input('Enter new PIN:')
        print('You PIN has been changed!!')

    def show_balance(self):
        print(f'''Your Current Balance is Rs {self.balance}''')

    def atm_performance_status(self):
        print('ATM working Fine 100%')
```

```python
icic = ATM()
```

```python
icic.show_balance()
```

```python
icic.balance
```

```python
icic.atm_performance_status()
```

```python
class ATM_version_1:
    def __init__(self):
        print('Object created successfully')
        self._pin = 1111   # protected variable
        self._balance = 2000  # protected variable
        print('*****')

    def change_pin(self):  # public method
        self._pin = input('Enter new PIN:')
        print('You PIN has been changed!!')

    def show_balance(self):
        print(f'''Your Current Balance is Rs {self._balance}''')

    def atm_performance_status(self):
        print('ATM working Fine 100%')
```

```python
axis = ATM_version_1()
```

```python
axis.show_balance()
```

```python
axis._pin
```

```python
class ATM_version_2:
    def __init__(self):
        print('Object created successfully')
        self.__pin = 1111   # private variable
        self.__balance = 2000  # private variable
        print('*****')

    def change_pin(self):  # public method
        self._pin = input('Enter new PIN:')
        print('You PIN has been changed!!')

    def show_balance(self):
        print(f'''Your Current Balance is Rs {self._balance}''')

    def __atm_performance_status(self): # private method
        print('ATM working Fine 100%')
```

```python
hdfc = ATM_version_2()
```

```python
hdfc._ATM_version_2__pin
```

```python
hdfc._ATM_version_2__atm_performance_status()
```

```python
##### Encapsulation in object-oriented programming (OOP) 
    # is the bundling of data (attributes) and methods (functions) 
    # that operate on that data into a single unit, usually a class. 
    # It also involves restricting direct access to some of the object's 
    # internal components, providing controlled access through methods, 
    # often referred to as getters and setters. 
    # This principle promotes data hiding, security, and code organization. 
```
