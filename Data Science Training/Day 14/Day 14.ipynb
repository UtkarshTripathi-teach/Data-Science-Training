```python
# Internet Required
df_titanic = sns.load_dataset('titanic')
df_titanic.sample()
```

```python
# sns.countplot
# sns.barplot
# sns.boxplot
# sns.scatterplot
# sns.catplot
# sns.pairplot
# sns.heatmap
# sns.regplot
# sns.lmplot
# sns.lineplot
# sns.kdeplot
# sns.violinplot
```

```python
import matplotlib.pyplot as plt

sns.countplot(data = df_titanic,x = 'sex', hue = 'alive', palette= sns.color_palette('magma',2))  # hue: further divide: based on categorical (object)
plt.title('Titanic survival Analysis Bases on gender!')
plt.show()
```

```python
# sns.barplot

sns.barplot(data = df_titanic, x='class',y='fare',hue = 'alive',estimator='mean') # min,max,sum
# Confidence Interval(CI): 95%
```

```python
# sns.barplot

sns.barplot(data = df_titanic, x='class',y='fare',hue = 'alive',errorbar=('ci',0),palette= sns.color_palette('cool',2)) 
plt.grid(color = 'r',alpha = 0.5)
plt.show()
# Confidence Interval(CI): 0%
```

```python
## sns.boxplot : To detect Outliers and to check data distribution
# To detectb Outliers: 
# Range: Min to Max (19-22): 35,28,45(Outliers: fliers))

# Q2: Quantile : 50%ile
# Q1: 25%ile
# Q3: 75%ile

## IQR = Q3 - Q1
# UW(upper cap) = Q3 + 1.5*IQR
# LW(lower cap) = Q1 - 1.5*IQR

```

```python
# sns.boxplot

sns.boxplot(data = df_titanic,x ='class',y = 'fare',hue='sex')
plt.show()
```

```python
# sns.boxplot

plt.figure(figsize=(10,5))
sns.boxplot(data = df_titanic,y ='class',x = 'fare',hue='sex',orient='h')

plt.show()
```

```python
# sns.violinplot

plt.figure(figsize=(10,5))
sns.violinplot(data = df_titanic,x ='class',y = 'fare',hue='sex',orient='v')

plt.show()
```

```python
df_tips = sns.load_dataset('tips')
df_tips.sample()
```

```python
# cat plot

sns.catplot(data = df_tips, x= 'time',y ='total_bill',kind='bar',hue='sex',row='smoker', col='day') # kind:"strip", "swarm", "box", "violin","boxen", "point", "bar", or "count".
plt.show()
```

```python
# cat plot

sns.catplot(data = df_tips, x= 'time',y ='total_bill',kind='violin',hue='sex',row='smoker', col='day') # kind:"strip", "swarm", "box", "violin","boxen", "point", "bar", or "count".
plt.show()
```

```python
# cat plot

sns.catplot(data = df_tips, x= 'total_bill',y ='tip',kind='point',hue='sex',row='smoker', col='day') # kind:"strip", "swarm", "box", "violin","boxen", "point", "bar", or "count".
plt.show()
```

```python
sns.lmplot(data=df_tips, x= 'total_bill', y='tip',row='smoker', col='day')
plt.show()
```

```python
sns.lmplot(data=df_tips, x= 'total_bill', y='tip',row='smoker', col='day',hue='sex')
plt.show()
```

```python
sns.relplot(data = df_tips, x= 'total_bill', y = 'tip',kind = 'scatter', hue= 'sex',row='smoker', col='day')
plt.show()
```

```python
# sns.pairplot(num vs num)

df_iris = sns.load_dataset('iris')
df_iris.sample(5)
```

```python
df_iris.iloc[:,:-1].corr()
```

```python
# >0: +ve : +ve Corr
# =0: No Corr
# <0: -ve : -ve Corr
```

```python
sns.pairplot(data = df_iris)
plt.show()
```

```python
sns.pairplot(data = df_iris , hue='species')
plt.show()
```

```python
sns.pairplot(data = df_iris , hue='species',corner=True)
plt.show()
```

```python
# sns.kdeplot
# sns.jointplot
```

```python
# sns.kdeplot : Kernal density estimation

sns.kdeplot(data = df_iris,x='sepal_length',y='petal_width',hue='species')
plt.show()
```

```python
# sns.kdeplot : Kernal density estimation

sns.kdeplot(data = df_iris,x='sepal_length',y='petal_width',hue='species')
sns.scatterplot(data = df_iris,x='sepal_length',y='petal_width',hue='species')
plt.show()
```

```python
# sns.kdeplot : Kernal density estimation

sns.kdeplot(data = df_iris,x='sepal_length',y='sepal_width',hue='species')
# sns.scatterplot(data = df_iris,x='sepal_length',y='sepal_width',hue='species')
plt.show()
```

```python
# sns.kdeplot : Kernal density estimation

sns.kdeplot(data = df_iris,x='sepal_length',y='sepal_width',hue='species')
sns.scatterplot(data = df_iris,x='sepal_length',y='sepal_width',hue='species')
plt.show()
```

```python
# sns.jointplot
```

```python
sns.jointplot(data = df_iris,x='sepal_length',y='sepal_width',hue='species')  
plt.show()
```

```python
sns.jointplot(data = df_iris,x='sepal_length',y='sepal_width',kind = 'hist')  # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```

```python
sns.jointplot(data = df_iris,x='sepal_length',y='sepal_width',kind = 'hex')  # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```

```python
sns.jointplot(data = df_iris,x='sepal_length',y='sepal_width',kind = 'reg')  # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```

```python
sns.jointplot(data = df_iris,x='sepal_length',y='sepal_width',kind = 'resid')  # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```
